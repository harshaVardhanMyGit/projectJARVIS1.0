```python
from dotenv import load_dotenv
from openai import OpenAI, AuthenticationError, RateLimitError, APIConnectionError
import os

load_dotenv()
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY1"))

def chatBot(question):
    try:
        completion = client.chat.completions.create(
            model = "gpt-4o-mini",
            messages = [
                {"role": "system", "content": "You are JARVIS, Tony Stark's sophisticated AI assistant. You are highly intelligent, polite, formal, and occasionally witty. Address the user as 'sir' or 'ma'am'. Provide precise, helpful responses with a touch of British elegance and subtle humor when appropriate."},
                {"role": "user", "content": question}
            ],
            max_tokens=150,
            temperature=0.3
        )
        return completion.choices[0].message.content
    
    except AuthenticationError:
        return "My apologies, sir. Authentication protocols have failed. Please verify your credentials."
    except RateLimitError:
        return "I'm afraid we've exceeded our operational limits, sir. Might I suggest we try again momentarily?"
    except APIConnectionError:
        return "It appears I'm experiencing connectivity issues, sir. Please check your network connection."
    except Exception as e:
        return f"Sir, I've encountered an unexpected error: {str(e)}. Shall I attempt to resolve this?"

if __name__ == "__main__":
    print("=" * 70)
    print("J.A.R.V.I.S. - Just A Rather Very Intelligent System")
    print("=" * 70)
    print("Good day, sir. All systems operational. How may I assist you today?")
    print("\nCommands: Type 'exit', 'quit', or 'bye' to terminate the session.\n")

    while True:
        question = input("Sir: ")
        if question.lower() in ['exit', 'quit', 'bye']:
            print("\nJARVIS: Very well, sir. Powering down. It's been a pleasure serving you.")
            break
        
        if not question.strip():
            print("JARVIS: I'm afraid I didn't catch that, sir. Please provide a valid query.")
            continue
        
        print("You: " + question)
        answer = chatBot(question)
        print(f"JARVIS: {answer}\n")
```
